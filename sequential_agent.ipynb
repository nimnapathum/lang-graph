```python
from typing import TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    name: str
    age: int
    final: str
```

```python
def first_node(state: AgentState) -> AgentState:
    """This is the first node in the sequence"""
    state['final'] = f"Hello {state['name']}, you are {state['age']} years old."
    return state

def second_node(state: AgentState) -> AgentState:
    """This is the second node in the sequence"""
    state['final'] += " Welcome to the LangGraph!"
    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node("first_node", first_node)
graph.add_node("second_node", second_node)
```

```python
graph.set_entry_point("first_node")
graph.add_edge("first_node", "second_node")
graph.set_finish_point("second_node")
app = graph.compile()
```

```python
app.invoke({'name': 'Alice', 'age': 30, 'final': ''})
```
